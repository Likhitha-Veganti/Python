```python
class Human:
    
    def __init__(self, name, age):
        self.__name = name
        self.__age = age  
        
    def hobby(self):
        print("Likes watching Netflix")  
        
    def info(self):
        print(self.__name,"is",self.__age)
        
```

```python
class Hiker(Human): 
    
    def __init__(self, name, age):
        Human.__init__(self,name,age)
    
    def hobby(self):
        print("Likes going on hikes")
        
```

```python
class Scientist(Human):
    
    def __init__(self, name, age, lab):
        Human.__init__(self, name, age)
        self.__lab = lab
        
    def hobby(self):
        print("Likes doing scientific experiments")
        self.labName()
        
    def labName(self):
        print("Works at",self.__lab,"lab")
```

```python
class Swimmer(Human):
    
    def __init__(self, name, age, hours ):
        Human.__init__(self, name, age)
        self.__hours = hours
        
    def hobby(self):
        print("Likes swimming")
        self.hoursSwimming()
        
    def hoursSwimming(self):
        print("Swims",self.__hours,"hours per week")
```

```python
class ScientificSwimmer(Scientist, Swimmer):
    
    def __init__(self, name, age, lab, hours):
        Scientist.__init__(self, name, age, lab)
        Swimmer.__init__(self, name, age, hours)
        
    def hobby(self):
        print("Likes swimming and doing scientific experiments")
        Scientist.labName(self)
        Swimmer.hoursSwimming(self)
```

```python
objects = []

l = ['1','2','3','4','5','q']

while(1):
    
    choice = input("1.Human, 2.Hiker, 3.Scientist, 4.Swimmer, 5.ScientificSwimmer, q.Exit...Enter your choice:\n")
    while(choice not in l):
        choice = input("Enter valid choice: ")
    if choice == 'q':
        break
    name = input("Enter the name: ")
    age = input("Enter the age: ")
    
    if choice == '1':
        obj = Human(name, age)
        objects.append(obj)
        
    elif choice == '2':
        obj = Hiker(name,age)
        objects.append(obj)
        
    elif choice == '3':
        lab = input("Enter the lab: ")
        obj = Scientist(name, age, lab)
        objects.append(obj)
    
    elif choice == '4':
        hours = input("Enter hours: ")
        obj = Swimmer(name, age, hours)
        objects.append(obj)
        
    elif choice == '5':
        lab = input("Enter the lab: ")
        hours = input("Enter hours: ")
        obj = ScientificSwimmer(name, age, lab, hours)
        objects.append(obj)
        
    print()
        
for obj in objects:
    print()
    obj.info()
    obj.hobby()
        
```
